We must note that in the autocorrelation plots presented above, the **dashed lines** are the approximate **two standard error confidence bounds** computed by ± 1,96* (1 / T), where T is the number observations.

If the autocorrelation is within these bounds, it is not significantly different from zero at (approximately)
5% level of significance.

So we conclude  that:

* From **ACF plot** we observe that we have **autocorrelation for lag=1 and lag=2**. We must check an autoregressive model for lag=2. **(AR(2))**
* From **PACF plot** we observe that we also have **partial autocorrelation for lag=1**. We must check a moving average model for lag=1. **(MA(1) )**

### UNIT ROOT TESTING

First, we can fit an **autoregressive time series model** to the Y2 data, by selecting the complexity of the
model **based on AIC**. Then, we perform an **augmented Dickey-Fuller test** of unit root, based on a model
with constant and trend 

```r
# Unit root testing for the Y1 series
# Fits an autoregressive model, and selects the best AR order based on AIC
m2 = ar(j2)
m2
paste('Order:', m2$order)
```

```r
m21 <- ur.df(j2, type='none', lags=m2$order-1)
m21
summary(m21)
```

##### Model withou drift and without trend
* The t-statistic is -5.4669 < -1.95 = Critical Value.
* Both z.lag.1 and z.diff.lag are statistically significant

```r
# Unit root testing for the Y1 series
# Fits an autoregressive model, and selects the best AR order based on AIC
m22 <- ur.df(j2, type="drift", lags=m2$order-1)
m22
summary(m22)
```

##### Model with drift and without trend
* The t-statistic is -7.0302 < -2.88 = Critical Value.
* Both Intercept and z.lag.1 are statistically significant
* z.diff.lag is not significant

```r
m23 <- ur.df(j2, type="trend", lags=m2$order-1)
m23
summary(m23)
```

##### Model with drift and trend
* The t-statistic is -7.2329 < -3.43 = Critical Value.
* Both Intercept and z.lag.1 are statistically significant
* z.diff.lag and tt (trend) are not significant

###   

We focus on the stationary time series, i.e. the monthly returns.
We can:
* **test for normality of Y2**
* **create the plot of the histogram together with a density plot**
* **also create the normal QQplot**

```r
# Shapiro test of normality
shapiro.test(y2)

par(mfrow=c(2,1))
hist(y2, prob=TRUE, 50) # histogram
lines(density(y2))
# smooth it - ?density for details
qqnorm(y2, main="Normal QQplot of Y2") # normal Q-Q plot
qqline(y2) # add a line
```

#### Conclusions:
* From Shapiro Wilk test (p-value= 1.751e-10) we **reject** the Null hypothesis for Normality.'
* From the Histogram we see a small deviance from Normality
* From the QQplot we see several points off the qqline.

So we conclude that we **don't have Normality**

### Estimation of ARMA models

#### We firstly check the moving average of lag=1 model (MA(1))

```r
ma1fit=arima(y2, order=c(0,0,1))
ma1fit
```

**The estimated model (MA(1) can be written in the form:**

$$y_{t} - 0.0084 = 0.2151* e_{t-1} + e_{t}$$

**or** 

$$y_{t} = 0.0084 + 0.2151* e_{t-1} + e_{t}$$


#### Now we will check the autoregressive of lag=1 model (AR(1))

```r
ar2fit <- arima(y2, order=c(2,0,0))
ar2fit
```

### The estimated model (AR(2) can be written in the form:

$$y_{t} = μ*(1-φ_{1}) + φ_{1} * y_{t-1} + φ_{2} * y_{t-2} + e_{t}$$

**or** 

$$y_{t} = 0.0062*(1- 0.2389) + 0.2389 * y_{t-1} + 0.1228 * y_{t-2} + e_{t}$$

**or** 

$$y_{t} = 0.00471882 + 0.2389 * y_{t-1} + 0.1228 * y_{t-2} + e_{t}$$


### Lets now fit an ARMA(2,1) model. The command is:

```r
arma21fit <- arima(y2, order=c(2,0,1))
arma21fit
```

#### The estimated ARMA(2, 1) model can be written in the form:


$$y_{t} - μ = φ_{1}*(y_{t-1} -μ) + φ_{2}*(y_{t-2} -μ) + θ_{1} * e_{t-1} + e_{t}$$

**or** 

$$y_{t} =μ +  φ_{1}*(y_{t-1} -μ) + φ_{2}*(y_{t-2} -μ) + θ_{1} * e_{t-1} + e_{t}$$

**or** 

$$y_{t} = 0.0062 +  0.0060*(y_{t-1} -0.0062) + 0.1850*(y_{t-2} -0.0062) +0.2369 * e_{t-1} + e_{t}$$


Summarizing we have the following:

### Diagnostic plots

Now we will provide some **diagnostic plots** for the **residuals** of the above models.
Based on the residuals of the AR(2) model, we will present the:
* autocorrelation **(acf)** plots and the partial autocorrelation **(pacf)** plots of the **estimated residuals**
* the autocorrelation plots and the partial autocorrelation plots of the **squared** residuals (check **heteroskedasticity**)
* normality plots (examine the assumption of **normality of residuals**).

```r
ar2residuals <- ar2fit$residuals
residuals <- ts(ar2residuals, frequency=12, start = c(1991, 1), end = c(2004, 12))

par(mfrow=c(3,2))

# set up the graphics
acf(ts(residuals,freq=1), 48, main="ACF of residuals")
pacf(ts(residuals,freq=1), 48, main="PACF of residuals")
acf(ts(residuals^2,freq=1), 48, main="ACF of squared residuals")
pacf(ts(residuals^2,freq=1), 48, main="PACF of squared residuals")
qqnorm(residuals,main="Normal QQplot of residuals")
qqline(residuals)
```
